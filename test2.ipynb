```python
import pandas as pd
import numpy as np
```

```python
def read_base(file_name):
    df=pd.read_csv(file_name, sep = ';', header=None, index_col = [1], parse_dates=[1])
    df.columns = ['Tag', 'Value', 'Quality']
    df.index.name='Date'
    df.sort_index(inplace=True)
    df=df[df['Quality'] == 'Good']
    return df
```

```python
test1=read_base('input/test1.txt')
test2=read_base('input/test2.txt')
```

```python
import plotly.graph_objects as naris
from plotly.offline import iplot, init_notebook_mode
init_notebook_mode(connected=True)
```

```python
fig = naris.Figure()
trace1 = naris.Scattergl(x=test1.index, y=test1['Value'], )
trace2 = naris.Scattergl(x=test2.index, y=test2['Value'], )
fig.add_trace(trace1)
fig.add_trace(trace2)
iplot(fig)
```
